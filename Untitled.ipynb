```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("paleo_data.csv")
```

```python
data.head()
```

My immediate reaction is that there are a lot of NaNs and a few columns which I'm not sure add value. Origin number and taxon number, as well as record type I'm not interested in. 

```python
missing_values_count = data.isnull().sum()
missing_values_count
```

I'm going to look into the NaN's a bit more and work out what percentage is missing. 

```python
total_cells = np.product(data.shape)
total_missing = missing_values_count.sum()
percentage_missing = (total_missing / total_cells) * 100
```

```python
print("The percentage of cells with missing data is " + str(round(percentage_missing)) + "%.")
```

I'm going to drop the columns which I don't need
